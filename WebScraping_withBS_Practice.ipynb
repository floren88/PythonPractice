# Demo: Web Scrapping with BeautifulSoup


```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.simplilearn.com/'
```

```python
result = requests.get(url)
webpage = result.content
sl_soup = BeautifulSoup(webpage,'html.parser')
```

```python
result.close()
```

```python
sl_soup.contents
```

```python
print(sl_soup.prettify())
```

```python
print(sl_soup.head)
```

```python
sl_soup.title
```

```python
# find all the links present on the web page
for href in sl_soup.findAll('a',href=True):
    print(href['href'])
```
