# Demo - Outliers and Missing Values


```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
dataset
```

```python
dataset.data
```

```python
dataset.target
```

```python
dataset['feature_names']
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(data=np.c_[dataset['data'],dataset['target']], columns = dataset['feature_names'] + ['target'])
```

```python
df
```

```python
df.isnull().any()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

for column in df:
    plt.figure()
    df.boxplot([column])
```
